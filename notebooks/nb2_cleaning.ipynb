```python
def fill_with_mode(df):
  # Handle potential empty column (avoiding errors)
  if df['subeixo_tecnologico_x'].dropna().empty:
      mode_value = None  # Set to None if col1 is empty
  else:
      mode_value = df['subeixo_tecnologico_x'].mode().iloc[0]  # Access the first mode value
  df['subeixo_tecnologico_x'] = df['subeixo_tecnologico_x'].fillna(mode_value)
  return df

# Apply the function to fill nulls
df = fill_with_mode(df.copy())
df['subeixo_tecnologico_x'].isna().sum()
```

### Modificação de dados

**Separa os tipos de curso em QUALIFICACAO, INTEGRADO, SUBSEQUENTE, CONCOMITANTE, EJA, SUPERIOR e POS_GRADUACAO**

```python
for df in cleaned:    
    df['novo_tipo_curso'] =  df.apply(
        lambda row: row['tipo_curso_x'] + '-' + row['tipo_oferta_x'] + '-' + row['modalidade_x'], 
        axis=1)
    print(df.columns)
```

```python
for df in cleaned:
    plt.figure()
    plt.bar(
        df['novo_tipo_curso'].value_counts().index, 
        df['novo_tipo_curso'].value_counts().values,
        color='purple'
    )
    plt.xlabel(col)
    plt.xticks(rotation=90)
    plt.title('Bar plot: novo_tipo_curso')
    plt.show()
```

```python
for df in cleaned:    
    # Cursos de Qualificação
   values = ['Qualificação Profissional (FIC)---Educação a Distância',
             'Qualificação Profissional (FIC)---Educação Presencial',
             'Qualificação Profissional (FIC)-Não se aplica-Educação a Distância',
             'Qualificação Profissional (FIC)-Não se aplica-Educação Presencial',
             'Qualificação Profissional (FIC)-Concomitante-Educação a Distância',
             'Qualificação Profissional (FIC)-Subsequente-Educação a Distância',
             'Qualificação Profissional (FIC)-Subsequente-Educação Presencial',
             'Qualificação Profissional (FIC)-Concomitante-Educação Presencial',
             'Qualificação Profissional (FIC)-Integrado-Educação a Distância',
             'Qualificação Profissional (FIC)-Integrado-Educação Presencial',
             'Qualificação Profissional (FIC)-PROEJA - Integrado-Educação Presencial',
             'Qualificação Profissional (FIC)-PROEJA - Concomitante-Educação Presencial',
             'Qualificação Profissional (FIC)-PROEJA - Concomitante-Educação a Distância',
             ]
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'QUALIFICACAO'
   
   ## Cursos Técnicos Integrados
   values = ['Técnico-Integrado-Educação Presencial']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'TECNICO_INTEGRADO'
   
   ## Cursos Técnicos Concomitantes
   values = ['Técnico-Concomitante-Educação Presencial','Técnico-Concomitante-Educação a Distância']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'TECNICO_CONCOMITANTE'
   
    ## Cursos Técnicos Subsequentes
   values = ['Técnico-Subsequente-Educação Presencial', 
             'Técnico-Subsequente-Educação a Distância']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'TECNICO_SUBSEQUENTE'

   ## Cursos Técnicos EJA
   values = ['Técnico-PROEJA - Integrado-Educação Presencial', 
             'Técnico-PROEJA - Concomitante-Educação Presencial']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'TECNICO_EJA'
   
   ## Cursos Superiores
   values = ['Tecnologia-Não se aplica-Educação Presencial',
             'Tecnologia-Não se aplica-Educação a Distância', 
             'Bacharelado-Não se aplica-Educação Presencial',
             'Tecnologia---Educação Presencial',
             'Tecnologia---Educação a Distância',
             'Bacharelado---Educação Presencial']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'SUPERIOR'
   
   ## Cursos de Pós-Graduação
   values = ['Especialização (Lato Sensu)-Não se aplica-Educação Presencial',
             'Especialização (Lato Sensu)-Não se aplica-Educação a Distância', 
             'Mestrado Profissional-Não se aplica-Educação Presencial',
             'Mestrado-Não se aplica-Educação Presencial',
             'Doutorado-Não se aplica-Educação Presencial',
             'Especialização (Lato Sensu)---Educação Presencial',
             'Especialização (Lato Sensu)---Educação a Distância', 
             'Mestrado Profissional---Educação Presencial',
             'Mestrado---Educação Presencial',
             'Doutorado---Educação Presencial']
   mask = df['novo_tipo_curso'].isin(values)
   df.loc[mask, 'novo_tipo_curso'] = 'POS_GRADUACAO'
```
