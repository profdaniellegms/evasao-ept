```python
Xet = X_categorical.drop(columns=['etnia', 'renda_familiar'])
yet = X_categorical['etnia']
```

```python
Xet_dum = pd.get_dummies(Xet)
Xet_dum
```

```python
Xet_join = Xet_dum.join(yet)
Xet_join
```

```python
Xet_split_y = Xet_join[Xet_join['etnia'].notnull()]
Xet_split_y
```

```python
Xet_split = Xet_split_y.drop('etnia', axis=1)
Xet_split
```

```python
yet
```

```python
yet_split = yet[yet.notnull()]
yet_split
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


Xet_train, Xet_test, yet_train, yet_test = train_test_split(Xet_split, yet_split, test_size=0.25, random_state=42)
```

```python
# Create the KNN classifier with k=5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)

# Train the KNN model on the training data
knn.fit(Xet_train, yet_train)

# Predict class labels for the test data
yet_pred = knn.predict(Xet_test)

# Evaluate the model performance (accuracy in this case)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(yet_test, yet_pred)
print(f"KNN Model Accuracy: {accuracy:.2f}")
```

### Label Encoding

```python
from sklearn.calibration import LabelEncoder


class MultiColumnLabelEncoder:

    def __init__(self, columns=None):
        self.columns = columns # array of column names to encode


    def fit(self, X, y=None):
        self.encoders = {}
        columns = X.columns if self.columns is None else self.columns
        for col in columns:
            self.encoders[col] = LabelEncoder().fit(X[col])
        return self


    def transform(self, X):
        output = X.copy()
        columns = X.columns if self.columns is None else self.columns
        for col in columns:
            output[col] = self.encoders[col].transform(X[col])
        return output


    def fit_transform(self, X, y=None):
        return self.fit(X,y).transform(X)


    def inverse_transform(self, X):
        output = X.copy()
        columns = X.columns if self.columns is None else self.columns
        for col in columns:
            output[col] = self.encoders[col].inverse_transform(X[col])
        return output
```

```python
multi = MultiColumnLabelEncoder(
    columns=['instituicao', 'regiao', 'estado', 'subeixo_tec', 'nome_curso', 'modalidade', 
             'turno', 'sexo', 'etnia', 'faixa_etaria', 'renda_familiar', 'tipo_curso'
        ]
    )

Xc = multi.fit_transform(X_categorical)
Xc_dcd = multi.inverse_transform(Xc)
```
